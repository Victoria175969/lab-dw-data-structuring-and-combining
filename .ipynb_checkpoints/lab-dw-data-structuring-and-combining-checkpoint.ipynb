In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df1 = pd.read_csv(url)
df1.rename(columns = {"ST": "State", "GENDER": "Gender"}, inplace=True)

# move column "Total Claim Amount" to the left 
column_to_move = "Total Claim Amount"
new_position = df1.columns.get_loc("Policy Type")  # Position before Policy Type

# Pop the column and insert it at the new position
df1.insert(new_position, column_to_move, df1.pop(column_to_move))
```

```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df2 = pd.read_csv(url)
df2.rename(columns = {"ST": "State", "GENDER": "Gender"}, inplace=True)
```

```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df3 = pd.read_csv(url)

# move column "Total Claim Amount" to the left 
column_to_move = "Total Claim Amount"
new_position1 = df3.columns.get_loc("Policy Type")  # Position before Policy Type
df3.insert(new_position1, column_to_move, df3.pop(column_to_move)) # Pop the column and insert it at the new position

# move column "Education" to the right 
column_to_move = "Education"
new_position3 = df3.columns.get_loc("Gender")  # Position before Income
df3.insert(new_position3, column_to_move, df3.pop(column_to_move))

# move column "Customer Lifetime Value" to the right 
column_to_move = "Customer Lifetime Value"
new_position2 = df3.columns.get_loc("Education")  # Position before Income
df3.insert(new_position2, column_to_move, df3.pop(column_to_move))
```

```python
df_new = pd.concat([df1, df2, df3], axis=0)
print(df_new.shape)
df_new
```

```python
df_new.dropna(how='all', inplace=True)
df_new.drop_duplicates(inplace=True)
print(df_new.shape)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
df = pd.read_csv(url)
df
```

```python
df[["sales_channel", "customer_lifetime_value", "total_claim_amount"]]
df["total_revenue"] = df[("customer_lifetime_value")] - df[("total_claim_amount")]
df["total_revenue"] = df["total_revenue"].round(2)
df[["sales_channel", "total_revenue"]]
total_revenue_sum = df["total_revenue"].sum()
print(total_revenue_sum)
df
```

```python
df[["sales_channel", "total_revenue"]]
```

```python
df_by_sales_channel = df[["sales_channel", "total_revenue"]]
df_by_sales_channel.groupby("sales_channel")["total_revenue"].sum()
new_df = df_by_sales_channel.groupby("sales_channel")["total_revenue"].sum()
new_df
```

```python
pivot_table = df.pivot_table(
        index = 'sales_channel',
        values='total_revenue', 
        aggfunc='sum')

pivot_table
```

```python
pivot_table = df.pivot_table(index='gender',
    values='customer_lifetime_value',
    columns='education',
    aggfunc='mean')   

print(" There is a negatiive correlation between the degree earned and average customer lifetime value, which is higher in the female group than in the male group.")          
pivot_table
```
